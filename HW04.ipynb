```python
import pandas as pd

import matplotlib.pyplot as plt
from statsmodels.graphics.gofplots import qqplot
from pandas.plotting import scatter_matrix

df = pd.read_csv("/content/housing.csv")

qqplot(df['AGE'], line='s')
plt.title("QQ Plot of Age")
plt.show()

correlation_matrix = df.corr()
scatter_matrix(correlation_matrix, alpha = 0.8, figsize=(10, 10), diagonal='hist')
```

```python
# Linear Regression

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error

X = df.drop('MEDV', axis=1)
y = df['MEDV']
lin_reg = LinearRegression()
lin_reg.fit(X, y)
y_pred = lin_reg.predict(X)

r2 = r2_score(y, y_pred)
mse = mean_squared_error(y, y_pred)

print('R-squared:', r2)
print('MSE:', mse)
print("\n")

plt.plot(y, color='blue', label='Actual Values')
plt.plot(y_pred, color='red', label='Predicted Values')
plt.xlabel('Index')
plt.ylabel('Value')
plt.title('Target vs. Predicted Values')
plt.legend()
plt.show()

print("\n")

residuals = y - y_pred
plt.plot(residuals, color='green')
plt.xlabel('Index')
plt.ylabel('Residuals')
plt.title('Residual Plot')
plt.show()
```

```python
# Ridge Regression

from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
import matplotlib.pyplot as plt
import numpy as np

train_df = df.sample(frac=0.8, random_state=42)
test_df = df.drop(train_df.index)

y_train = train_df['MEDV'].values
X_train = train_df.drop('MEDV', axis=1).values
y_test = test_df['MEDV'].values
X_test = test_df.drop('MEDV', axis=1).values

model = Ridge()


alpha_values = [0.01, 0.1, 1, 5, 10, 20, 50, 100, 1000]

param_grid = {'alpha': alpha_values}
grid = GridSearchCV(model, param_grid, cv=5)
grid.fit(X_train, y_train)

print('Best alpha:', grid.best_params_['alpha'])
print('Best score:', grid.best_score_)


best_alpha = grid.best_params_['alpha']
model = Ridge(alpha=best_alpha)
model.fit(X_train, y_train)


y_pred = model.predict(X_test)


r2 = model.score(X_test, y_test)
mse = np.mean((y_test - y_pred)**2)
print('R2:', r2)
print('MSE:', mse)
print("\n")

plt.plot(y_test, color='blue', label='Actual Values')
plt.plot(y_pred, color='red', label='Predicted Values')
plt.xlabel('Index')
plt.ylabel('Value')
plt.title('Target vs. Predicted Values')
plt.legend()
plt.show()

print("\n")

residuals = y_test - y_pred
plt.plot(residuals, color='green')
plt.xlabel('Index')
plt.ylabel('Residuals')
plt.title('Residual Plot')
plt.show()
```

```python
#lasso regression

from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
import matplotlib.pyplot as plt
import numpy as np


train_df = df.sample(frac=0.8, random_state=42)
test_df = df.drop(train_df.index)

y_train = train_df['MEDV'].values
X_train = train_df.drop('MEDV', axis=1).values
y_test = test_df['MEDV'].values
X_test = test_df.drop('MEDV', axis=1).values

model = Lasso()

alpha_values = [0.01, 0.1, 1, 5, 10, 20, 50, 100, 1000]

param_grid = {'alpha': alpha_values}
grid = GridSearchCV(model, param_grid, cv=5)
grid.fit(X_train, y_train)

print('Best alpha:', grid.best_params_['alpha'])
print('Best score:', grid.best_score_)

best_alpha = grid.best_params_['alpha']
model = Lasso(alpha=best_alpha)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

r2 = model.score(X_test, y_test)
mse = np.mean((y_test - y_pred)**2)
print('R2:', r2)
print('MSE:', mse)

print("\n")

plt.plot(y_test, color='blue', label='Actual Values')
plt.plot(y_pred, color='red', label='Predicted Values')
plt.xlabel('Index')
plt.ylabel('Value')
plt.title('Target vs. Predicted Values')
plt.legend()
plt.show()

print("\n")

residuals = y_test - y_pred
plt.plot(residuals, color='green')
plt.xlabel('Index')
plt.ylabel('Residuals')
plt.title('Residual Plot')
plt.show()
```
