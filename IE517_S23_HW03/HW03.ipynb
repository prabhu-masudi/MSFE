```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

bond_yields = pd.read_csv("/content/HY_Universe_corporate bond.csv")

lower_whisker = bond_yields["Coupon"].quantile(0.25) - 1.5 * (bond_yields["Coupon"].quantile(0.75) - bond_yields["Coupon"].quantile(0.25))
upper_whisker = bond_yields["Coupon"].quantile(0.75) + 1.5 * (bond_yields["Coupon"].quantile(0.75) - bond_yields["Coupon"].quantile(0.25))
median = bond_yields["Coupon"].median()
lower_quartile = bond_yields["Coupon"].quantile(0.25)
upper_quartile = bond_yields["Coupon"].quantile(0.75)

plt.boxplot(bond_yields["Coupon"], whis=[lower_whisker, upper_whisker], showfliers=False)

plt.title("Distribution of Bond Yields")
plt.xlabel("Coupon")
plt.ylabel("Bond")
plt.plot([1, 1], [lower_quartile, upper_quartile], color='red', linewidth=2)
plt.plot([1, 1], [median, median], color='green', linewidth=2)

plt.show()
```

```python
plt.scatter(bond_yields["bond_type"], bond_yields["Coupon"])

plt.title("Coupon Yield vs. Bond Type")
plt.xlabel("Bond Type")
plt.ylabel("Coupon Yield")
plt.ylim(0,20)

plt.show()
```

```python
plt.scatter(bond_yields["bond_type"], bond_yields["LIQ SCORE"])

plt.title("Liquid score vs. Bond Type")
plt.xlabel("Bond Type")
plt.ylabel("Liquid score")
plt.ylim(0,1)

plt.show()
```

```python
plt.scatter(bond_yields["n_trades"], bond_yields["volume_trades"])

plt.title("Number of Trades vs. Volume of Trades")
plt.xlabel("Number of trades")
plt.ylabel("Volume of trades")
plt.show()
```

```python
plt.hist(bond_yields["volume_trades"], bins=50)

plt.title("Distribution of Volume of Trades")
plt.xlabel("Volume of Trades")
plt.ylabel("Frequency")

plt.show()
```

```python
grouped = bond_yields.groupby("bond_type").mean()

grouped["volume_trades"].plot(kind="bar")

plt.title("Mean Volume of Trades by Bond Type")
plt.xlabel("Bond Type")
plt.ylabel("Mean Volume of Trades")

plt.show()
```
